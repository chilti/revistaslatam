```
import pandas as pd

#Camiar nombre de archivo

fileName = 'MAtrix evol Mex smooth.csv'

Data = pd.read_csv(fileName, sep=',', index_col = 0)
Data.head()
```

```
Data.plot(figsize=(18, 10), title="Data Mining")
```

```
# Generar dataframe suavisado con 3 a√±os
### Cambiar el p√°metro para una media m√≥vil mayor

Data_Smooth = Data.copy()
for column in Data.columns:
  Data_Smooth[column] =  Data[column].rolling(3, min_periods=1, win_type='exponential', center=True).mean(tau=1)
Data_Smooth.plot(figsize=(18, 10), title='Data Analysis')
```

```
Data_Smooth.to_csv('Smoothed_'+fileName, sep=';')
```
